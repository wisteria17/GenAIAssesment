```python
# ==========================================
# 1. Import Libraries
# ==========================================
!pip install PyPDF2
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ==========================================
# 2. (Optional) PDF Resume Reader
# ==========================================
import PyPDF2

def extract_text_from_pdf(file_path):
    text = ""
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

# 👉 Uncomment below if using PDF
# resume = extract_text_from_pdf("resume.pdf")

# ==========================================
# 3. Sample Resume & Job Description
# ==========================================
resume = """
Skilled in Python, Machine Learning, Data Analysis, Pandas, NumPy.
Experience in building ML models and data visualization.
"""

job_description = """
Looking for a candidate with Python, Machine Learning, SQL, Deep Learning skills.
Experience in Data Analysis is required.
"""

# ==========================================
# 4. Text Preprocessing
# ==========================================
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

resume_clean = preprocess(resume)
jd_clean = preprocess(job_description)

# ==========================================
# 5. TF-IDF + Cosine Similarity
# ==========================================
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([resume_clean, jd_clean])

similarity_score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
match_percentage = round(similarity_score * 100, 2)

# ==========================================
# 6. Skill-Based Tracing
# ==========================================
skills_list = [
    "python", "machine learning", "data analysis", "sql",
    "deep learning", "nlp", "pandas", "numpy",
    "data visualization", "tensorflow", "keras"
]

def extract_skills(text):
    found_skills = []
    for skill in skills_list:
        if skill in text:
            found_skills.append(skill)
    return set(found_skills)

resume_skills = extract_skills(resume_clean)
jd_skills = extract_skills(jd_clean)

matched_skills = resume_skills.intersection(jd_skills)
missing_skills = jd_skills - resume_skills

# ==========================================
# 7. Decision Logic
# ==========================================
if match_percentage >= 60:
    decision = "Selected"
else:
    decision = "Rejected"

# ==========================================
# 8. Output Results
# ==========================================
print("===== AI Resume Screening Result =====\n")

print(f"Match Score: {match_percentage}%")
print(f"Decision: {decision}\n")

print("Matched Skills:")
if matched_skills:
    for skill in sorted(matched_skills):
        print(f"✔ {skill}")
else:
    print("None")

print("\nMissing Skills:")
if missing_skills:
    for skill in sorted(missing_skills):
        print(f"✘ {skill}")
else:
    print("None")

# ==========================================
# 9. Explanation (Tracing)
# ==========================================
print("\nExplanation:")
if decision == "Selected":
    print("Candidate has most of the required skills and matches the job description well.")
else:
    print("Candidate is missing important skills required for the job.")
```
